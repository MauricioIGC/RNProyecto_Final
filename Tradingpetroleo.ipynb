# Predice usando el dato de cierre

```python
import numpy as np
np.random.seed(4)
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM as tf


#Precio de Petróleo crudo Brent desde 02/02/2010 hasta 02/02/2022
dataBrent = pd.read_csv('PetroleoBrent.csv')
#Precio de Petróleo crudo WTI desde 02/02/2010 hasta 02/02/2022
dataWTI = pd.read_csv('PetroleoWTI.csv')

dataBrent.head()
```

```python
#Arreglando lo de la fecha y poniendolo como indice
dataBrent['Fecha'] = pd.to_datetime(dataBrent['Fecha'], format="%d.%m.%Y")
dataWTI['Fecha'] = pd.to_datetime(dataWTI['Fecha'], format="%d.%m.%Y")

#Poniendo en órden (el último elemento es el último día del dataset)
dataBrent = dataBrent.sort_values('Fecha')
dataWTI = dataWTI.sort_values('Fecha')

len(dataBrent)
```

```python
#Poniendo las fechas en los índices
dataBrent.index=dataBrent["Fecha"]
dataWTI.index=dataWTI["Fecha"]
dataBrent.head()
```

```python
#Gráficas de Brent y WTI
fig, ax = plt.subplots()
ax.set_xlabel('Tiempo')
ax.set_ylabel('Cierre')
ax.plot(dataBrent['Fecha'],dataBrent['Cierre'], 'b')
ax.plot(dataWTI['Fecha'],dataWTI['Apertura'], 'r')
plt.show()
```

```python
#Dividiendo en conjunto de entrenamiento y validación para únicamente el petróleo Brent
set_entrena = dataBrent.iloc[:2842,1:2]
set_validac = dataBrent.iloc[2842:,1:2]

#Normalizando el set de entrenamiento
uno = MinMaxScaler(feature_range=(0,1))
set_entrena_norm = uno.fit_transform(set_entrena)
len(set_entrena_norm)
```

```python
# Poniendo los datos en la forma en q se entrenarán, osea en un num de datos consecutivos (60), y de salida 1 dato (la predicción a
# partir de esos datos),ie se dividirá en bloques de 60 el conjunto de entrenamiento (0-59)(1-60),(2-61)..
datos_consec = 60
X_train = []
Y_train = []
m = len(set_entrena_norm)

for i in range(datos_consec,m):
    # X: bloques de 60 datos: 0-59, 1-60, 2-61, etc
    X_train.append(set_entrena_norm[i-datos_consec:i,0])

    # Y: el siguiente dato
    Y_train.append(set_entrena_norm[i,0])
    
X_train, Y_train = np.array(X_train), np.array(Y_train)

# Ajustando al modelo en Keras
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
#Recuerda checar que sí coincida este Y_train con el de más abajo, sino el de abajo esta mal
Y_train
```

```python

# Red LSTM

dim_entrada = (X_train.shape[1],1)
dim_salida = 1
na = 60

modelo = Sequential()
modelo.add(tf(units=na, input_shape=dim_entrada))
modelo.add(Dense(units=dim_salida))
modelo.compile(optimizer='rmsprop', loss='mse')
modelo.fit(X_train,Y_train,epochs=100,batch_size=32)
```

```python
# Validación (predicción del valor de las acciones)
#
x_test = set_validac.values
x_test = uno.transform(x_test)

X_test = []
for i in range(datos_consec,len(x_test)):
    X_test.append(x_test[i-datos_consec:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

prediccion = modelo.predict(X_test)
prediccion = uno.inverse_transform(prediccion)

```

```python
# Graficar resultados
#graficar_predicciones(set_validac.values,prediccion)

plt.plot(set_validac.values[60:len(prediccion)+60],color='red', label='Valor del petróleo Brent')
plt.plot(prediccion, color='blue', label='Predicción')
plt.ylim(1.1 * np.min(prediccion)/2, 1.1 * np.max(prediccion))
plt.xlabel('Tiempo')
plt.ylabel('Valor del petróleo')
plt.legend()
plt.show()

#fig, ax = plt.subplots()
#ax.set_xlabel('Tiempo')
#ax.set_ylabel('Apertura)')
#ax.plot(dataBrent['Fecha'],dataBrent['Apertura'], 'b')
#ax.plot(dataWTI['Fecha'],dataWTI['Apertura'], 'r')
#plt.show()
```

# A partir de aquí es teniendo en cuenta el precio WTI

```python
#Dividiendo en conjunto de entrenamiento y validación para los 2 petróleos
set_entrenaB = dataBrent.iloc[:2842,1:2]
set_validacB = dataBrent.iloc[2842:,1:2]
set_entrenaW = dataWTI.iloc[:2842,1:2]
set_validacW = dataWTI.iloc[2842:,1:2]


uno = MinMaxScaler(feature_range=(0,1))
Brent_entrena_norm = uno.fit_transform(set_entrenaB)

dos = MinMaxScaler(feature_range=(0,1))
WTI_entrena_norm = dos.fit_transform(set_entrenaW)
Brent_entrena_norm
```

```python
#El set que usaremos en el entrenamiento será una combinación de estos sets
norm_fin = []
for j in range(0,2*len(WTI_entrena_norm)):
    if (j%2==0) :
        #En los pares ponemos el Brent
        norm_fin.append(Brent_entrena_norm[int(j/2)])
    else:
        norm_fin.append(WTI_entrena_norm[int((j-1)/2)])
norm_final=np.asarray(norm_fin)
len(norm_final)
```

```python
# Poniendo los datos en la forma en q se entrenarán, osea en un num de datos consecutivos (120), y de salida 1 dato (la predicción a
# partir de esos datos),ie se dividirá en bloques de 120 el conjunto de entrenamiento (0-119),(2-121)..
datos_consec = 120
X_train = []
Y_train = []
m = len(norm_final)

for i in range(datos_consec,m):
    if i%2==0:
        # X: bloques de 120 datos
        X_train.append(norm_final[i-datos_consec:i,0])

        # Y: el siguiente dato
        Y_train.append(norm_final[i,0])
    
X_train, Y_train = np.array(X_train), np.array(Y_train)

# Ajustando al modelo en Keras
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

Y_train
```

```python
dim_entrada = (X_train.shape[1],1)
dim_salida = 1
na = 60

modelo2 = Sequential()
modelo2.add(tf(units=na, return_sequences=True, input_shape=dim_entrada))
modelo2.add(tf(units=40, return_sequences=False))
modelo2.add(Dense(units=dim_salida))
modelo2.compile(optimizer='rmsprop', loss='mse')
modelo2.fit(X_train,Y_train,epochs=100,batch_size=32)
```

```python
Brent_test_norm = uno.fit_transform(set_validacB)
WTI_test_norm = dos.fit_transform(set_validacW)
#convirtiendo el dataset de validación en uno como el de entrenamiento

validac_fin = []
for j in range(0,520):
    if (j%2==0) :
        #En los pares ponemos el Brent
        validac_fin.append(Brent_test_norm[int(j/2)])
    else:
        validac_fin.append(WTI_test_norm[int((j-1)/2)])
validac_final=np.asarray(validac_fin)


# Validación (predicción del valor del petroleo)
x_test = validac_final

X_test = []
for i in range(datos_consec,len(x_test)):
    X_test.append(x_test[i-datos_consec:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

prediccion = modelo2.predict(X_test)
#Para cambiar entre predecir el precio de Brent por WTI hay que cambiar al "uno.inv.." por "dos.inv.." y cosas de más abajo
prediccionF = uno.inverse_transform(prediccion)
len(prediccionF)
```

```python
prediccionFin= np.zeros(200)
for i in range(0,len(prediccionFin)):
#Para cambiar entre predecir el precio de Brent por WTI hay que cambiar al [(i*2)] por [(i*2)+1] y el plot para verlo
    prediccionFin[i]=prediccionF[(i*2)]
len(prediccionFin)
```

```python
# Graficar resultados
#Para cambiar entre predecir el precio de Brent por WTI hay que cambiar al set_validacB por set_validacW
plt.plot(set_validacB.values[59:len(prediccionFin)+59],color='red', label='Valor del petróleo Brent')
plt.plot(prediccionFin, color='blue', label='Predicción')
plt.ylim(1.1 * np.min(prediccionFin)/2, 1.1 * np.max(prediccionFin))
plt.xlabel('Tiempo')
plt.ylabel('Valor deL petróleo')
plt.legend()
plt.show()

#fig, ax = plt.subplots()
#ax.set_xlabel('Tiempo')
#ax.set_ylabel('Apertura)')
#ax.plot(dataBrent['Fecha'],dataBrent['Apertura'], 'b')
#ax.plot(dataWTI['Fecha'],dataWTI['Apertura'], 'r')
#plt.show()
```
